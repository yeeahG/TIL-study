### 데이터 추출

```python
mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')
```

```python
df[mask].shape
```

```python
df_seoul = df[mask]
df_seoul
```

```python
#df_seoul = df_seoul.drop(['전출지별'], axis=1) # 전출지별 삭제
```

```python
df_seoul = df_seoul.rename(columns = {'전입지별' : '전입지'}) # rename
```

```python
df_seoul
```

```python
df_seoul = df_seoul.set_index('전입지') # set_index
```

```python
sr_one = df_seoul.loc['경기도']
```

```python
sr_one
```

## 시각화 1

```python
plt.plot(sr_one.index, sr_one.values)
```

```python
from matplotlib import font_manager, rc
```

```python
font_path = "./data/malgun.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
print(plt.style.available)
```

```python
# 스타일
plt.style.use('seaborn-muted')

# 그림 크기
plt.figure(figsize=(14,5)) #14인치 5인치

# x 축의 각도
#plt.xticks(rotation='vertical')
plt.xticks(rotation=45)

plt.title("서울 -> 경기 인구 이동", fontdict = {'fontsize' : 15})

# 축 이름 추가
plt.xlabel('기간')
plt.ylabel('인구 이동수')

plt.plot(sr_one)

#Legend
plt.legend(labels=['서울 -> 경기'], loc='best', fontsize=15)
# loc='lower right' # 안나오다가 위치 잘넣어서 나옴

# 축의 범위 ylim xlim
plt.ylim(0, 700000)

# 주석 표시 - 화살표
plt.annotate('',
             xy=(20, 620000),       #화살표의 머리 부분(끝점)
             xytext=(2, 290000),    #화살표의 꼬리 부분(시작점)
             xycoords='data',       #좌표체계
             arrowprops=dict(arrowstyle='->', color='skyblue', lw=5), #화살표 서식
             )

plt.annotate('',
             xy=(47, 450000),       #화살표의 머리 부분(끝점)
             xytext=(30, 580000),    #화살표의 꼬리 부분(시작점)
             xycoords='data',       #좌표체계
             arrowprops=dict(arrowstyle='->', color='olive', lw=5), #화살표 서식
             )

plt.annotate("인구이동 증가",
            xy=(10,450000),
            rotation=22,
            va='baseline', # 텍스트 상하 정렬
            ha='center',   # 텍스트 좌우 정렬
            fontsize=15
            )

plt.annotate("인구이동 감소",
            xy=(39,550000),
            rotation=-8,
            va='baseline', # 텍스트 상하 정렬
            ha='center',   # 텍스트 좌우 정렬
            fontsize=15
            )

plt.show()
```
