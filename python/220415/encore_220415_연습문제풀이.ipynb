## 문제 4
### 위의 결과를 바탕으로 목적지 공항 별 결항 횟수, 회항 횟수
### 운항 횟수를 구하시오 (Cancelled, Diverted, Air)
##### 운항 횟수는 결항과 회항을 제외한 것 

```python
hflight2.groupby(['Dest']).count()
```

```python
hflight2[hflight2['Dest'] == 'ATL']['DepTime'].isnull().sum()
```

```python
cnt = hflight2.groupby(['Dest'])[['Year']].count()
```

```python
cnt
```

```python
hflight2['Cancelled'].value_counts()
```

```python
Cancelled = hflight2.groupby(['Dest'])[['Cancelled']].sum()
```

```python
Cancelled
```

```python
Cancelled['Cancelled'].sum()
```

```python
hflight2['Diverted'].value_counts()
```

```python
Diverted = hflight2.groupby(['Dest'])[['Diverted']].sum()
```

```python
Diverted
```

```python
Diverted['Diverted'].sum()
```

```python
df4 = pd.concat([Cancelled, Diverted, cnt], axis=1)
```

```python
df4
```

```python
df4['Air'] = df4.apply(lambda x : x[2] - x[0] - x[1], axis=1)
```

```python
df4
```

```python
df4['Air'] = df4['Year'] - df4['Cancelled'] - df4['Diverted']
```

```python
df4
```

```python
df4.drop("Year", axis=1, inplace=True)
```
